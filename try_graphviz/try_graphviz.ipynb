```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from matplotlib import pyplot as plt

# to use graphviz
import pydotplus as pdp
from sklearn.tree import export_graphviz
from IPython.display import Image

% matplotlib inline
```

```python
iris = load_iris()
iris.keys()
```

```python
df_all = pd.DataFrame(iris['data'],columns=iris['feature_names'])
df_all.head()
```

```python
# (versicolor, not versicolor) <-> (1,0)
df_all['target_' + str(iris['target_names'][1])]=[i if i==1 else 0 for i in iris['target']]
df_all.head()
```

```python
train, test = train_test_split(df_all, stratify=df_all['target_versicolor'],test_size=0.3)
print(train.shape,test.shape)
```

### build models and search parmeter

```python
%%time

#  parmeter
upper_depth = 10
depths = range(1, upper_depth)

# score
auc_scores = []

# model
models = []

# calc
for depth in depths:
    clf = DecisionTreeClassifier(max_depth=depth)
    clf.fit(train[iris['feature_names']],train['target_versicolor'])
    
    predicted_label = clf.predict(test[iris['feature_names']])
    fpr, tpr, thresholds = metrics.roc_curve(test['target_versicolor'].values,predicted_label)
    auc_scores.append(metrics.auc(fpr,tpr))
    
    models.append(clf)
```

```python
auc_scores
```

### display graph of tree

```python
best_model = 2

export_graphviz(models[best_model], out_file='tree.dot', feature_names=iris['feature_names'], class_names=['0','1'], filled=True, rounded=True)
graph = pdp.graph_from_dot_file(path='tree.dot')
Image(graph.create_png())
```

```python
pd.DataFrame(np.array([iris['feature_names'],models[best_model].feature_importances_]).T,columns=['feature','importance'])
```
